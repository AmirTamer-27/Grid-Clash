```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Baseline test
client1 = pd.read_csv("./Baseline/baseline_client_metrics_0.csv")
client2 = pd.read_csv("./Baseline/baseline_client_metrics_1.csv")
client3 = pd.read_csv("./Baseline/baseline_client_metrics_2.csv")
client4 = pd.read_csv("./Baseline/baseline_client_metrics_3.csv")
serverMetrics = pd.read_csv("./Baseline/baseline_server_metrics.csv")
# --- PLOT 1: LATENCY OVER TIME ---
plt.figure(figsize=(10, 5))
plt.plot(client1['time_since_start_ms']/1000, client1['latency_ms'], label='Latency', color='blue', linewidth=0.8)
plt.title('Client Latency Over Time')
plt.xlabel('Time (seconds)')
plt.ylabel('Latency (ms)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.savefig('latency_graph.png')
plt.show()


mc1 = client1["latency_ms"].mean()
mc2 = client2["latency_ms"].mean()
mc3 = client3["latency_ms"].mean()
mc4 = client4["latency_ms"].mean()
avgLatency = (mc1+mc2+mc3 +mc4)/4
print(str(mc1))
print(str(mc2))
print(str(mc3))
print(str(mc4))
print("Average Latency: " + str(avgLatency))
print(serverMetrics)
```

```python
# 2% packet loss
client1 = pd.read_csv("./Loss_2%/loss_2%_client_metrics_0.csv")
client2 = pd.read_csv("./Loss_2%/loss_2%_client_metrics_1.csv")
client3 = pd.read_csv("./Loss_2%/loss_2%_client_metrics_2.csv")
client4 = pd.read_csv("./Loss_2%/loss_2%_client_metrics_3.csv")

mc1 = client1["latency_ms"].mean()
mc2 = client2["latency_ms"].mean()
mc3 = client3["latency_ms"].mean()
mc4 = client4["latency_ms"].mean()
avgLatency = (mc1+mc2+mc3 +mc4)/4

mc1p = client1["perceived_position_error"].mean()
mc2p = client2["perceived_position_error"].mean()
mc3p = client3["perceived_position_error"].mean()
mc4p = client4["perceived_position_error"].mean()

p95_val1 = client1['perceived_position_error'].quantile(0.95)
p95_val2 = client1['perceived_position_error'].quantile(0.95)
p95_val3 = client1['perceived_position_error'].quantile(0.95)
p95_val4 = client1['perceived_position_error'].quantile(0.95)
avg95 =(p95_val1+p95_val2+p95_val3 +p95_val4)/4
avgMPE = (mc1p+mc2p+mc3p +mc4p)/4
# --- PLOT 1: LATENCY OVER TIME ---
plt.figure(figsize=(10, 5))
plt.plot(client1['time_since_start_ms']/1000, client1['latency_ms'], label='Latency', color='blue', linewidth=0.8)
plt.title('Client Latency Over Time')
plt.xlabel('Time (seconds)')
plt.ylabel('Latency (ms)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.savefig('latency_graph.png')
plt.show()

plt.figure(figsize=(10, 5))
plt.plot(client1['time_since_start_ms']/1000, client1['perceived_position_error'], label='perceived_position_error', color='blue', linewidth=0.8)
plt.title('perceived_position_error over Time')
plt.xlabel('Time (seconds)')
plt.ylabel('perceived_position_error')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
print("Average mean percieved error position : " + str(avgMPE))
print("95th percentile mpe: " + str(avg95))
print("Average Latency: " + str(avgLatency))

```

```python
# 5% packet loss
client1 = pd.read_csv("./Loss_5%/loss_5%_client_metrics_0.csv")
client2 = pd.read_csv("./Loss_5%/loss_5%_client_metrics_1.csv")
client3 = pd.read_csv("./Loss_5%/loss_5%_client_metrics_2.csv")
client4 = pd.read_csv("./Loss_5%/loss_5%_client_metrics_3.csv")
total_packets = len(client1)
success_packets = len(client1[client1['latency_ms'] <= 200])
reliability_score = (success_packets / total_packets) * 100
print(f"Reliability: {reliability_score}%") 

mc1 = client1["latency_ms"].mean()
mc2 = client2["latency_ms"].mean()
mc3 = client3["latency_ms"].mean()
mc4 = client4["latency_ms"].mean()
avgLatency = (mc1+mc2+mc3 +mc4)/4

mc1p = client1["perceived_position_error"].mean()
mc2p = client2["perceived_position_error"].mean()
mc3p = client3["perceived_position_error"].mean()
mc4p = client4["perceived_position_error"].mean()
avgMPE = (mc1p+mc2p+mc3p +mc4p)/4
# --- PLOT 1: LATENCY OVER TIME ---
plt.figure(figsize=(10, 5))
plt.plot(client1['time_since_start_ms']/1000, client1['latency_ms'], label='Latency', color='blue', linewidth=0.8)
plt.title('Client Latency Over Time')
plt.xlabel('Time (seconds)')
plt.ylabel('Latency (ms)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.savefig('latency_graph.png')
plt.show()

print("Average mean percieved error position : " + str(avgMPE))
print("Average Latency: " + str(avgLatency))

```
